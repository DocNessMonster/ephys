```python
import pandas as pd

# Specify the encoding if the default 'utf-8' doesn't work
encoding = 'latin1'  # You can also try 'iso-8859-1' if 'latin1' doesn't work

# Read the CSV file with the specified encoding
df = pd.read_csv('HS23.csv', encoding=encoding)

# Filter rows where both 'pF' and 'Size' columns have data
filtered_df = df.dropna(subset=['pF', 'Size'])

# Set 'Abf' column as the index (row labels)
labeled_df = filtered_df.set_index('Abf')

# Select only the 'pF' and 'Size' columns
result_df = labeled_df[['pF', 'Size']]

# Save the result to a new CSV file if needed
result_df.to_csv('HS23_extracted.csv')

print(result_df)
```
